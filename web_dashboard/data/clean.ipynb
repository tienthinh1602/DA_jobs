```python
import numpy as np
import pandas as pd
import re

def clean_salary_range(x):
    # Handle NaN values
    if pd.isna(x):
        return np.nan
    
    # Convert to string
    x = str(x).lower().replace(' ', '')
    
    # Remove Vietnamese currency and extra characters
    x = x.replace('tr', '').replace('triệu', '').replace('vnđ/tháng', '')
    
    # Special cases
    if x in ['khôngtiếtlộ', 'vílídobaomậtnênkhôngtiếtlộ']:
        return np.nan
    
    # Handle comparison ranges
    if '>' in x:
        x = x.replace('>', '').strip()
        try:
            val = float(x)
            return '40-100' if val >= 30 else np.nan
        except:
            return np.nan
    
    if '<' in x:
        x = x.replace('<', '').strip()
        try:
            val = float(x)
            return '0-5' if val <= 5 else np.nan
        except:
            return np.nan
    
    # Remove any double hyphens or extra symbols
    x = x.replace('--', '-')
    
    # Try to extract numeric ranges
    match = re.findall(r'(\d+)[-]?(\d+)?', x)
    
    if match:
        start = float(match[0][0])
        end = float(match[0][1]) if match[0][1] else start
        
        # Standardize ranges
        if end <= 5:
            return '0-5'
        elif 5 < end <= 10:
            return '5-10'
        elif 10 < end <= 20:
            return '10-20'
        elif 20 < end <= 30:
            return '20-30'
        elif 30 < end <= 40:
            return '30-40'
        elif end > 40:
            return '40-100'
    
    # If no match found, try direct numeric conversion
    try:
        val = float(x)
        if val <= 5:
            return '0-5'
        elif 5 < val <= 10:
            return '5-10'
        elif 10 < val <= 20:
            return '10-20'
        elif 20 < val <= 30:
            return '20-30'
        elif 30 < val <= 40:
            return '30-40'
        elif val > 40:
            return '40-100'
    except:
        return np.nan
    
    return np.nan

result['Khoảng lương hiện tại'] = result['Khoảng lương hiện tại'].apply(clean_salary_range)
result['Khoảng lương của công việc đầu tiên'] = result['Khoảng lương của công việc đầu tiên'].apply(clean_salary_range)
```
